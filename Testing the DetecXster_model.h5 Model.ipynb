```python
## Required imports...
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import load_model
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
## Loading the Trainned "DetecXster_model.h5" model for testing.
cnn = load_model("C:/Users/SEAN/Desktop/data/DetecXster_model.h5")
```

```python
## Path for the test Folder.
test_path = "C:/Users/SEAN/Desktop/data/test"
```

```python
## Saving all the images inside the test folder to a list. And Displaying all the images in the order.
test_x = []
i= 1

for img in os.listdir(test_path):
    ##print(img)
    
    ## Setting the image path.
    image_path = test_path+"/"+img
    ## Loading the Image
    img_pic = image.load_img(image_path, target_size=(256,256,3))
    
    ## Setting the image to plot.
    fig = plt.figure(figsize=(30, 15))
    fig.add_subplot(3, 9, i)
    plt.imshow(img_pic)
    plt.axis('off')
    plt.title(str(i-1)+ " = " + img)
    
    i = i + 1
    
    ## Normalizing the data.
    img_pic = image.img_to_array(img_pic) / 255
    
    ##img = np.expand_dims(img,axis=0)
    
    ## Store the image in the list
    test_x.append(img_pic)  
```

```python
## Changing the list into a numpy array.
test_x = np.array(test_x)
```

```python
## Predicting using the Model.
result = cnn.predict(test_x)
```

```python
## Storing the predicted indexes inside a list.
predicted_index = [np.argmax(i) for i in result]
```
